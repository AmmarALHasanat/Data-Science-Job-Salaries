```python
comp_loc = ds_salaries['company_location'].unique()
```

```python
years = ds_salaries['work_year'].unique()
exp = ds_salaries['experience_level'].unique()
emp = ds_salaries['employment_type'].unique()
job = ds_salaries['job_title'].unique()
comp_size = ds_salaries['company_size'].unique()
curr = ds_salaries['salary_currency'].unique()
```

```python
df_each_loc_list = []


for loc in comp_loc:
    df_each_loc = pd.DataFrame()
    
    df_loc = ds_salaries.loc[ds_salaries['company_location']==loc]
    max_salary = df_loc['salary_in_usd'].max()
    
    df_each_loc['Year'] = df_loc['work_year'].loc[df_loc['salary_in_usd']==max_salary].values
    df_each_loc['Company_location'] = loc
    df_each_loc['Max. salary (in USD)'] = max_salary
    
    df_each_loc_list.append(df_each_loc)

df_loc = pd.concat(df_each_loc_list , axis=0)

df_loc.head(3)
```

```python
df_loc_2020 = df_loc[['Company_location','Max. salary (in USD)']].loc[df_loc['Year']==2020]
df_loc_2021 = df_loc[['Company_location','Max. salary (in USD)']].loc[df_loc['Year']==2021]
df_loc_2022 = df_loc[['Company_location','Max. salary (in USD)']].loc[df_loc['Year']==2022]
```

```python
fig, axes = plt.subplots(3, 1 , figsize=(14,7))
axes[0].bar(df_loc_2020['Company_location'] , df_loc_2020['Max. salary (in USD)'] , label='2020' , color = 'r')
axes[1].bar(df_loc_2021['Company_location'] , df_loc_2021['Max. salary (in USD)'] , label='2021' , color='b')
axes[2].bar(df_loc_2022['Company_location'] , df_loc_2022['Max. salary (in USD)'] , label='2022' , color='g')

plt.suptitle('Company location vs Maximum salary over 3 years')
fig.supylabel('Maximum salary (in USD)')
fig.supxlabel('Company location')
fig.legend()
plt.show()
```

```python
df_each_job_list = []


for j in job:
    df_each_job = pd.DataFrame()
    
    df_job = ds_salaries.loc[ds_salaries['job_title'] == j ]
    max_salary = df_job['salary_in_usd'].max()
    
    df_each_job['Year'] = df_job['work_year'].loc[df_job['salary_in_usd']==max_salary].values
    df_each_job['Job_title'] = j
    df_each_job['Max. salary (in USD)'] = max_salary
    
    df_each_job_list.append(df_each_job)

df_job = pd.concat(df_each_job_list , axis=0)

df_job_2020 = df_job[['Job_title','Max. salary (in USD)']].loc[df_job['Year']==2020]
df_job_2021 = df_job[['Job_title','Max. salary (in USD)']].loc[df_job['Year']==2021]
df_job_2022 = df_job[['Job_title','Max. salary (in USD)']].loc[df_job['Year']==2022]

fig, axes = plt.subplots(3, 1 , figsize=(20,9))
axes[0].bar(df_job_2020['Job_title'] , df_job_2020['Max. salary (in USD)'] , label='2020' , color = 'r' , width = 0.3)
axes[1].bar(df_job_2021['Job_title'] , df_job_2021['Max. salary (in USD)'] , label='2021' , color='b' , width=0.3)
axes[2].bar(df_job_2022['Job_title'] , df_job_2022['Max. salary (in USD)'] , label='2022' , color='g' , width=0.3)


plt.suptitle('Job-titles vs Maximum salary over 3 years', fontsize = 15)
fig.supylabel('Maximum salary (in USD)' , fontsize=12)
fig.supxlabel('Job-title' , fontsize=12)
fig.legend()
plt.show()
```

```python
df_each_exp_list = []


for e in exp:
    df_each_exp = pd.DataFrame()
    
    df_exp = ds_salaries.loc[ds_salaries['experience_level'] == e ]
    max_salary = df_exp['salary_in_usd'].max()
    
    df_each_exp['Year'] = df_exp['work_year'].loc[df_exp['salary_in_usd']==max_salary].values
    df_each_exp['Experience_level'] = e
    df_each_exp['Max. salary (in USD)'] = max_salary
    
    df_each_exp_list.append(df_each_exp)

df_exp = pd.concat(df_each_exp_list , axis=0)

df_exp_2020 = df_exp[['Experience_level','Max. salary (in USD)']].loc[df_exp['Year']==2020]
df_exp_2021 = df_exp[['Experience_level','Max. salary (in USD)']].loc[df_exp['Year']==2021]
df_exp_2022 = df_exp[['Experience_level','Max. salary (in USD)']].loc[df_exp['Year']==2022]

fig, axes = plt.subplots(3, 1 , figsize=(10,4))
axes[0].bar(df_exp_2020['Experience_level'] , df_exp_2020['Max. salary (in USD)'] , label='2020' , color = 'r' , width = 0.3)
axes[1].bar(df_exp_2021['Experience_level'] , df_exp_2021['Max. salary (in USD)'] , label='2021' , color='b' , width=0.3)
axes[2].bar(df_exp_2022['Experience_level'] , df_exp_2022['Max. salary (in USD)'] , label='2022' , color='g' , width=0.3)


plt.suptitle('Experience-levels vs Maximum salary over the years', fontsize = 15)
fig.supylabel('Maximum salary (in USD)' , fontsize=12)
fig.supxlabel('Experience-levels' , fontsize=12)
fig.legend()
plt.show()
```
